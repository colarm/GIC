### 2.2 Use a for loop to calculate the BMI of each person.
>$\frac{Weight(kg)}{Height(m)^2} = BMI$  

```python
### YOUR CODE GOES HERE ###
BMI_list = []
for item in person_data:
    BMI_list.append([item[0], item[1] / (item[2] / 100) ** 2])
BMI_list
```

### 2.2 BONUS: Do it using list comprehension.

```python
### YOUR CODE GOES HERE ###
BMI_list = []
for item in person_data:
    BMI_list.append([item[0], item[1] / (item[2] / 100) ** 2])
BMI_list
```
