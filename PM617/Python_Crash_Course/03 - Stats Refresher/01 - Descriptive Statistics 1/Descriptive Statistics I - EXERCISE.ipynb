### Visualize the distribution of tips. Describe it (skew, kurtosis, etc.).

```python

### YOUR CODE GOES HERE ####
from scipy.stats import kurtosis, skew
plt.style.use('ggplot')
df['tip'].hist(bins='auto')

print(kurtosis(df["tip"]))
print(skew(df["tip"]))
```
